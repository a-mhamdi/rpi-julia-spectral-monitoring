**Type hierarchy**

Julia's numeric types form a tree. `supertype` walks up; `subtypes` walks down:

````julia
typeof(3.14)
````

````julia
# Walk up from Float64 to Any
Float64 |> supertype |> supertype |> supertype |> supertype
````

````julia
# All concrete integer subtypes
Integer |> subtypes
````

````julia
Signed |> subtypes
````

````julia
UInt8 <: Unsigned   # true — UInt8 is a subtype of Unsigned
Signed >: Int8      # true — Signed is a supertype of Int8
````

**Optional positional arguments (splat operator)**

Default values make arguments optional. Use `...` to splat a collection into positional arguments:

````julia
foo(x=0, y=0, z=0) = x + y + z
````

````julia
foo(), foo(1, 2, 3)
````

````julia
foo([1, 2, 3]...)   # splat unpacks the vector into x, y, z
````

**Keyword arguments**

Keyword arguments are declared after `;` in the signature. They can also be splatted from a `Dict`:

````julia
bar(; a::Real=0, b::Real=0, c::Real=0) = a + b + c
````

````julia
bar()
````

````julia
bar(; Dict(:a => 3, :b => 5.4, :c => -1.2)...)
````

Positional splatting does **not** work for keyword arguments — the following raises a `MethodError`:

````julia
try
    bar([1, 2, 3]...)
catch err
    println(err)
end
````

**Multiple Dispatch**

Julia selects which method to call based on the types of *all* arguments — not just the first. Each specialization is called a *method*:

````julia
# Method 1: integer argument
function f(x::Int)
    x^2
end

# Method 2: Float64 argument
f(x::Float64) = x^2 + 1

# Method 3: Char argument
f(x::Char) = string(x, 'y', 'z')

# Method 4: String argument
f(x::String) = x * x
````

````julia
methods(f)   # list all defined methods
````

````julia
f(1), f(1.0), f('x'), f("abc")
````

**Extending an existing function** — adding a second method to `mycos`:

````julia
mycos(x) = cos(x)
mycos(adj, hyp) = adj / hyp   # geometric interpretation: cosine = adjacent / hypotenuse
````

````julia
methods(mycos)
````

````julia
@which mycos(π)        # which method gets called?
````

````julia
@which mycos(5, 3)
````

Adding an optional argument updates the dispatch table:

````julia
mycos(adj, hyp=10) = adj / hyp
````

````julia
@which mycos(π)
````

**Function Chaining**

The pipe operator `|>` passes the result of one expression as the argument to the next function:

````julia
g(x) = x + 1
h(x) = x^2
x = 2 |> g |> h   # equivalent to h(g(2))
````

The composition operator `∘` (*\circ + Tab*) creates a new function without immediately calling it:

````julia
(h ∘ g)(2)   # same result as above
````

Anonymous functions work inline in a pipeline:

````julia
y = 5 |> (x -> x^2) |> √
````

**Metaprogramming & Introspection**

Julia compiles code to native machine code via LLVM. You can inspect the generated IR with `@code_llvm`:

````julia
function compute_sum_of_squares(x::Integer)
    y = x
    for i in 1:100
        y += i^2
    end
    return y
end
````

````julia
@code_llvm compute_sum_of_squares(3)
````

````julia
?@code_llvm
````

### Logical Arrays

Boolean vectors can be used as index masks to select elements:

````julia
x = [1, 2, -5, 7.2, 3im]
println(x)
println(typeof(x))
````

````julia
idx = [false, true, false, false, true]
println(x[idx])   # selects elements at positions 2 and 5
````

Boolean masks also work on matrices for row/column selection:

````julia
M = Array{Float64, 2}(undef, 5, 4)
````

````julia
row_idx = [true, false, true, true, false]
col_idx = [false, true, true, false]
````

````julia
M[row_idx, :]           # selected rows, all columns
````

````julia
M[:, col_idx]           # all rows, selected columns
````

````julia
M[row_idx, col_idx]     # selected rows and columns
````

### Control Flow

**Conditional evaluation:**

````julia
a, b = 3.0, 5.0
if a < b
    println("$a is less than $b")
elseif a > b
    println("$a is greater than $b")
else
    println("$a equals $b")
end
````

**`while` loop:**

````julia
fruits = ["Blueberry", "Orange", "Banana", "Raspberry", "Strawberry"]
iter = 1
while iter ≤ length(fruits)
    println("Item #$iter is $(fruits[iter])")
    iter += 1
end
````

**`for` loop** — iterating directly over a collection is idiomatic Julia. Use `enumerate` to get both the index and the value:

````julia
vegetables = ["Broccoli", "Garlic", "Mushrooms", "Potatoes"]
for (i, item) in enumerate(vegetables)
    println("Item #$i is $item")
end
````

### Final Project: Basic Calculator

Bring together what you've learned by building a simple calculator. It should:
1. Accept two numbers `x`, `y` and a `Char` operation.
2. Return the result of the chosen arithmetic operation.
3. Raise a meaningful error for unsupported operators.

*Extensions to try:* handle more operations (square root, power, trig), accept keyword-argument style input, or build an interactive REPL loop.

````julia
function calculator(x::Number, y::Number, op::Char)
    if op == '+'
        return x + y
    elseif op == '-'
        return x - y
    elseif op == '*'
        return x * y
    elseif op ∈ ('/', '÷')
        y == 0 && error("Division by zero")
        return x / y
    else
        error("Unsupported operator: '$op'")
    end
end
````

````julia
println("5 + 3 = $(calculator(5, 3, '+'))")
println("5 - 3 = $(calculator(5, 3, '-'))")
println("5 × 3 = $(calculator(5, 3, '*'))")
println("5 ÷ 3 = $(calculator(5, 3, '÷'))")

# Test error handling
try
    calculator(5, 3, '×')
catch err
    println("Error: ", err.msg)
end
````

### Modules

Modules create isolated namespaces. `export` makes names available after `using`:

````julia
module MyModule
    export a          # `a` is public; `b` is internal
    a = 0
    b = true
end
````

````julia
varinfo(MyModule)   # inspect module contents
````

Access names with `ModuleName.name`. After `using .MyModule`, exported names are available directly:

````julia
MyModule.a, MyModule.b   # always accessible via qualified name
````

````julia
using .MyModule
a             # available because it was exported
````

Import only a specific name with `using .MyModule: name`:

````julia
using .MyModule: b   # explicitly import the unexported `b`
b
````
